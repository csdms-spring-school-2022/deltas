# Case Study: Wax Lake Delta

## Characterizing Streamflow

Wax Lake is located at...  The USGS has a streamgage... we will use the `nwis` package to retrieve the streamflow data and characterize its geomorphic work

```python
#import packages
import dataretrieval.nwis as nwis
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#get water discharge data from usgs
site = '07381590'

# get instantaneous values (iv)
df = nwis.get_record(sites=site, service='dv', start='1986-06-01', end='2015-09-30')
```

```python
#load data

#get dates
df['date'] = df.index
#convert water discharge from ft^3/s to m^3/s
df['q'] = df['00060_Mean']*0.0283168
#find year
df['yr'] = pd.DatetimeIndex(df['date']).year
```

```python
#compute flow frequency curve

q_flood = np.round(df.groupby('yr')['q'].agg({np.nanmax}))
q_f = -np.sort(-q_flood['nanmax'])
t_flood = np.arange(1, len(q_flood)+1)
T = (len(q_flood)+1)/t_flood
# P = 1/T
# G = q_f*P   
# PT = 1-P**100

return_period = 1.5

#id of the max value
flood_id = df.groupby('yr')['q'].idxmax()
```

```python
q_fo = np.interp(return_period, np.flip(T), np.flip(q_f))

plt.loglog(T, q_f, '-.ko', linewidth = 0.5)
plt.plot(return_period, q_fo, 'ro')
plt.axhline(y = q_fo, color = 'r', linestyle = '-', alpha=0.5, 
            label= 'Q$_{' + str(return_period) + ' year}$ = ' + str(round(q_fo)) +' [m$^3$/s]')
plt.axvline(x = return_period, color = 'r', linestyle = '-', alpha=0.5)
plt.ylabel('Flood Water Discharge [m$^3$/s]')
plt.xlabel('Recurrence Interval [yr]' )
plt.tight_layout()
# plt.xlim([0.5, 110])
plt.xlim([0.9,40])
plt.legend()
plt.show()
```

```python
#plot
plt.plot(df['date'], df['q'], 'b')
plt.plot(flood_id, q_flood, 'r*', label = 'annual max. water discharge')
plt.axhline(y = q_fo, color = 'r', linestyle = '-.', alpha=0.5, 
            label= 'Q$_{' + str(return_period) + ' year}$ = ' + str(round(q_fo)) +' [m$^3$/s]')
plt.xlabel('Year')
plt.ylabel('Flood Water Discharge [m$^3$/s]')
plt.ylim([0 ,1e4])
plt.legend()
plt.show()
```

```python
#intermittency 

df['qr'] = df['q']/q_fo
plt.plot(df['date'], df['qr'], 'b')
plt.xlabel('Year')
plt.axhline(y = 1, color = 'r', linestyle = '-.', alpha=0.5, 
            label= 'Q$_{' + str(return_period) + ' year}$ = ' + str(round(q_fo)) +' [m$^3$/s]')
plt.ylabel('Intermittency Factor')
plt.legend()
plt.ylim([0,2])
plt.show()
```

```python
If = df.groupby('yr')['qr'].apply(lambda x: x[x >= 1].count())/365.5
sns.histplot(x=If,stat='probability', color = 'b', alpha = 0.4)
plt.axvline(x = If.mean() , color = 'b', linestyle = '-',  label = 'mean = ' + str(np.round(If.mean(), 3)))
plt.axvline(x = If.median() , color = 'b', linestyle = '-.', label = 'median = ' + str(np.round(If.median(), 3)))
plt.xlabel('Intermittency Factor')
plt.rcParams["patch.force_edgecolor"] = False
plt.legend()
plt.show()

print('Average flood duration is ', (If.mean()*365.5), 'days')
print('Median flood duration is ', (If.median()*365.5), 'days')
```

```python
plt.figure()
yr_uni = np.unique(df['yr'])
plt.plot(yr_uni, If, '-bo')
plt.ylabel('Intermittency Factor')
plt.xlabel('year')
plt.show()
```
